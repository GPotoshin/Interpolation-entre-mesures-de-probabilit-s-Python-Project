**Montrons que $\mathcal{W}_p(\mu_0, \mu_1) \ll \infty$ si $\mu_0, \mu_1 \in \mathcal{P}_p(\mathbb{R})$.**

Montrons que $\forall a, b \in \mathbb{R}$, $ |a - b|^p \leq |a|^p + |b|^p$.

**Initialisation**

Pour $p = 1$, inégalité triangulaire suffit pour conclure.

**Hypothèse** : la propriété est vérifiée pour p. Montrons la pour p+1.

$$
 |a - b|^{p+1} = |a - b| \cdot |a - b|^{p} \leq (|a| + |b|) \cdot (|a|^{p} + |b|^{p}) \leq  |a|^p + |b|^p
$$

Donc $\forall u \in [0,1]$,
$$
 |\mathcal{F}_0^{-1}(u) - \mathcal{F}_1^{-1}(u)|^p \leq |\mathcal{F}_0^{-1}(u)|^p + |\mathcal{F}_1^{-1}(u)|^p
$$

En intégrant,
$$
\int_0^1 |\mathcal{F}_0^{-1}(u) - \mathcal{F}_1^{-1}(u)|^p \, du \leq \int_0^1 |\mathcal{F}_0^{-1}(u)|^p \, du + \int_0^1 |\mathcal{F}_1^{-1}(u)|^p \, du
$$

On effectue le changement de variables :
$x = \mathcal{F}_0^{-1}(u)$, donc $du = d\mu_0(x)$,
$y = \mathcal{F}_1^{-1}(u)$, donc $du = d\mu_1(y)$.

Donc
$$
\mathcal{W}_p(\mu_0, \mu_1) \leq \int_{\mathbb{R}} |x|^p \, d\mu_0(x) + \int_{\mathbb{R}} |y|^p \, d\mu_1(y).
$$

Comme $\mu_0, \mu_1 \in \mathcal{P}_p(\mathbb{R})$, on a $\mathcal{V}_p(\mu_0, \mu_1) \ll \infty$.



**Montrons que $\mathcal{W}_p$ est une distance sur $\mathcal{P}_p(\mathbb{R})$**
$$
Soit \mu_0,\mu_1,\mu_2 \in \mathcal{P}_p(\mathbb{R}
$$
- **Symétrie** :
On a $\mathcal{W}_p(\mu_0, \mu_1) = \mathcal{W}_p(\mu_1, \mu_0)$ d'après les propriété de la valeur absolue.

- **Séparation** :
  Soit $\mu_0, \mu_1 \in \mathcal{P}_p(\mathbb{R})$.
  Si $\mu_0 = \mu_1$, alors $\mathcal{F}_0^{-1} = \mathcal{F}_1^{-1}$ donc $\mathcal{W}_p(\mu_0, \mu_1) = 0$.

  Si $\mathcal{W}_p(\mu_0, \mu_1) = 0$, alors
  $$
  \int_0^1 |\mathcal{F}_0^{-1}(u) - \mathcal{F}_1^{-1}(u)|^p \, du = 0.
  $$
  Comme $|\mathcal{F}_0^{-1}(u) - \mathcal{F}_1^{-1}(u)|^p \geq 0$, on a $|\mathcal{F}_0^{-1}(u) - \mathcal{F}_1^{-1}(u)| = 0$ p.p.
  Donc $\mathcal{F}_0^{-1} = \mathcal{F}_1^{-1}$ par continuité, ce qui implique $\mu_0 = \mu_1$.

- **Inégalité triangulaire** :
  Soit $\mu, \mu_0, \mu_1 \in \mathcal{P}_p(\mathbb{R})$.
  On a
  $$
 |\mathcal{F}_\mu^{-1}(u) - \mathcal{F}_{\mu_1}^{-1}(u)|^p \leq |\mathcal{F}_\mu^{-1}(u) - \mathcal{F}_{\mu_0}^{-1}(u)|^p + |\mathcal{F}_{\mu_0}^{-1}(u) - \mathcal{F}_{\mu_1}^{-1}(u)|^p.
  $$
  En intégrant,
  $$
  \mathcal{W}_p(\mu, \mu_1) \leq \mathcal{W}_p(\mu, \mu_0) + \mathcal{V}_p(\mu_0, \mu_1).
  $$

On détermine la fonction de répartition de $\mu_0$ :

Soit $x' \in \mathbb{R}$ :
$$
F_0(x') = \begin{cases}
0 & \text{si } x' < x_{(1)}, \\
\frac{k}{N} & \text{si } x' \in [x_{(k)}, x_{(k+1)}[ \text{ pour } k \in [[1, N-1]], \\
1 & \text{si } x' \geq x_{(N)}.
\end{cases}
$$

Ainsi, la fonction quantile inverse de $F_0$ est :

Soit $u \in [0, 1[$,
$$
F_0^{-1}(u) = x_{(k)} \quad \text{si } u \in \left[\frac{k-1}{N}, \frac{k}{N}\right[ \text{ avec } k \in [[1, N]].
$$

On a la même expression pour $\mu_1$ avec les $y_{(k)}$.

Donc
$$
\begin{aligned}
\mathcal{W}_p^p(\mu_0, \mu_1) &= \int_0^1 |F_0^{-1}(u) - F_1^{-1}(u)|^p \, du \\

&= \sum_{k=1}^N \frac{1}{N} |x_{(k)} - y_{(k)}|^p.
\end{aligned}
$$

$$
\mathcal{W}_p^p(\mu_0, \mu_1) = \frac{1}{N} \sum_{k=1}^N |x_{(k)} - y_{(k)}|^p.
$$
